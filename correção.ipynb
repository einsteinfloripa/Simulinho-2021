# Métricas

## Alunos

```python
# calcula o total de questões certas por máteria, quantas questões tinham por matéria, e média por matéria

# agg permite agregar multiplas funções ao groupby
acertos_aluno = data.groupby(['Nome','Matéria'])['Correção'].agg(['sum','count','mean'])
acertos_aluno = acertos_aluno.rename(columns={"sum": "Total Acertos", "mean": "Média Individual", 'count': "Total questoes"})
```

```python
acertos_aluno.describe()
```

```python
# comparar a média de acerto das questões dos alunos com a turma geral

# media geral do einstein
media_acerto_materia = data.groupby('Matéria').mean()

media_acerto_materia = media_acerto_materia.rename(columns={"Correção": "Media Geral"})


# adicionando a media geral aos dados individuais
acertos_aluno = acertos_aluno.join(media_acerto_materia, on='Matéria')
acertos_aluno
```

```python
pontuacao_total = acertos_aluno.groupby("Nome").sum()
pontuacao_total["Total questoes"] = 47
pontuacao_total
```

```python
pontuacao_total.describe()
```

```python
# identificar alunos que não fizeram o simuliinho
pontuacao_total["Zerou"] = pontuacao_total["Total Acertos"] == 0
```

```python
acertos_aluno = acertos_aluno.unstack()
acertos_aluno.head()
```

```python
acertos_aluno[('Total Acertos', 'Total')] = acertos_aluno['Total Acertos'].T.sum()
```

```python
acertos_aluno[('Total questoes', 'Total')] = 47
```

```python
acertos_aluno[('Média Individual', 'Total')] = acertos_aluno[('Total Acertos', 'Total')] / 47
```

```python
acertos_aluno
```

```python
acertos_aluno = acertos_aluno.stack()
media_acertos_obj = data['Correção'].mean()
acertos_aluno = acertos_aluno.fillna(value=media_acertos_obj)
acertos_aluno
```

```python
acertos_aluno.xs('Total', level='Matéria')
```

```python
# acertos_aluno[acertos_| acertos_aluno["Total Acertos"] == 0]
```

```python
# manipulando a coluna de total de acertos pra mostrar no formato -> acertos/total de questoes
acertos_aluno[['Total Acertos', 'Total questoes']] = acertos_aluno[['Total Acertos', 'Total questoes']].astype(str)

acertos_aluno['Total de Acertos'] = acertos_aluno[['Total Acertos', 'Total questoes']].agg('/'.join, axis=1)

acertos_aluno = acertos_aluno.drop(columns =['Total Acertos', 'Total questoes'])
```

```python
# reordenando as colunas
acertos_aluno = acertos_aluno[['Total de Acertos',  "Média Individual", "Media Geral"]]
acertos_aluno
```

```python
# salvando como pickle para o relatorio
acertos_aluno.to_pickle('dados-relatorio-alunos/acertos_aluno.pkl')
```

```python
# calculando a colocação dos alunos
colocacao = pontuacao_total.sort_values(by=['Total Acertos'], ascending=False)

# como os indices ficam fora de ordem depois de colocar as notas em valor decrescente, é necessário resetar o valor do indices
# de 'colocacao'
colocacao = colocacao.reset_index()

colocacao['Colocação'] = colocacao.index + 1
colocacao["% de acerto"] = round((colocacao["Total Acertos"] / colocacao["Total questoes"])*100, 3)
# salvando como pickle para o relatorio
colocacao.to_pickle('dados-relatorio-alunos/colocacao.pkl')


colocacao
```
